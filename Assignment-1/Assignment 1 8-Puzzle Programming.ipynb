```python
import numpy as np 
import queue 
import time 
GOAL = [0, 1, 2, 3, 4, 5, 6, 7, 8]

class Puzzle():
    def __init__(self,current_state,parent_node,action_state,tree_depth,step_cost_factor,path_cost_factor):
        self.current_state = current_state 
        self.parent_node = parent_node 
        self.action_state = action_state 
        self.tree_depth = tree_depth 
        self.step_cost_factor = step_cost_factor 
        self.path_cost_factor = path_cost_factor        
        self.shift_up = None 
        self.shift_left = None
        self.shift_down = None
        self.shift_right = None
        step_counter = 0
        global counter 
    
    def move_right(self):
        initial_state_index=[i[0] for i in np.where(self.current_state==0)] 
        if initial_state_index[1] == 0:
            return False
        else:
            move_left_index_value = self.current_state[initial_state_index[0],initial_state_index[1]-1] 
            new_state = self.current_state.copy()
            new_state[initial_state_index[0],initial_state_index[1]] = move_left_index_value
            new_state[initial_state_index[0],initial_state_index[1]-1] = 0
            return new_state,move_left_index_value
        
    def move_up(self):
        initial_state_index=[i[0] for i in np.where(self.current_state==0)] 
        if initial_state_index[0] == 2:
            return False
        else:
            lower_index_value = self.current_state[initial_state_index[0]+1,initial_state_index[1]] # value of the lower tile
            new_state = self.current_state.copy()
            new_state[initial_state_index[0],initial_state_index[1]] = lower_index_value
            new_state[initial_state_index[0]+1,initial_state_index[1]] = 0
            return new_state,lower_index_value
        
    def move_down(self):
        initial_state_index=[i[0] for i in np.where(self.current_state==0)] 
        if initial_state_index[0] == 0:
            return False
        else:
            move_up_index_value = self.current_state[initial_state_index[0]-1,initial_state_index[1]] 
            new_state = self.current_state.copy()
            new_state[initial_state_index[0],initial_state_index[1]] = move_up_index_value
            new_state[initial_state_index[0]-1,initial_state_index[1]] = 0
            return new_state,move_up_index_value
        
    def move_left(self):
        initial_state_index=[i[0] for i in np.where(self.current_state==0)] 
        if initial_state_index[1] == 2:
            return False
        else:
            move_right_value = self.current_state[initial_state_index[0],initial_state_index[1]+1] # value of the move_right tile
            new_state = self.current_state.copy()
            new_state[initial_state_index[0],initial_state_index[1]] = move_right_value
            new_state[initial_state_index[0],initial_state_index[1]+1] = 0
            return new_state,move_right_value
        
    def path_trace(self):
        global counter
        current_state_trace = [self.current_state]
        action_state_trace = [self.action_state]
        tree_depth_trace = [self.tree_depth]
        step_cost_factor_trace = [self.step_cost_factor]
        path_cost_factor_trace = [self.path_cost_factor]     
        
        while self.parent_node:
            self = self.parent_node
            current_state_trace.append(self.current_state)
            action_state_trace.append(self.action_state)
            tree_depth_trace.append(self.tree_depth)
            step_cost_factor_trace.append(self.step_cost_factor)
            path_cost_factor_trace.append(self.path_cost_factor)            
        step_counter = 0
        print(self.current_state)
        while current_state_trace:         
            print('to')
            print (current_state_trace.pop())            
            step_counter += 1     
        counter = step_counter  
        print(counter)
```
